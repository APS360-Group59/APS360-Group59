## 0. Imports

```
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
```

## 1. Loading Data

```
audio_path = '/content/Smash Mouth - All Stars (Lyrics) ðŸŽµ (320  kbps).mp3'
x , sr = librosa.load(audio_path, sr = 44100)
```

```
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

## 2. Effects and Data Augmentation Testing

```
array_1, array_2 = librosa.effects.hpss(x)
```

```
ipd.Audio(array_1, rate = sr)
```

```
ipd.Audio(array_2, rate = sr)
```

```
pitch_shift = librosa.effects.pitch_shift(x, sr, 50)
ipd.Audio(pitch_shift, rate = sr)
```

```
bad_mouth = np.clip(x, 0., 1.)
ipd.Audio(bad_mouth, rate = sr)
```

```
pitch_shift = librosa.effects.pitch_shift(x, sr, -50)
ipd.Audio(pitch_shift, rate = sr)
```
